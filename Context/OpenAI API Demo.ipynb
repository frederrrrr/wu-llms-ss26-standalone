## 3.3 Approximate cost calculation (per API call, 2.000 output tokens)
Note: 2000 tokens ≈ 1500–1600 words ≈ 3 full pages of text

```python
from openai import OpenAI
client = OpenAI(api_key="INSERT-YOUR-API-KEY-HERE")

prompt = "Write a bedtime story about a flying unicorn."

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[{"role": "user", "content": prompt}],
    max_completion_tokens= 2000
)

input_token_price = 0.40 / 1_000_000
output_token_price = 1.60 / 1_000_000


# Extract token usage
max_completion_tokens= 2000
input_tokens = response.usage.prompt_tokens
output_tokens = max_completion_tokens
# Calculate cost
cost = (input_tokens * input_token_price + output_tokens *  output_token_price)
print(f"Estimated cost: ${cost}")
```

## 3.4 Load input
Csv file incluing prompts: https://huggingface.co/datasets/fka/awesome-chatgpt-prompts/blob/main/prompts.csv

```python
import pandas as pd

URL = "https://huggingface.co/datasets/fka/awesome-chatgpt-prompts/resolve/main/prompts.csv" #download URL of CSV

df = pd.read_csv(URL)
df = df.dropna(subset=["prompt"]).reset_index(drop=True)
df.head(3)
```

```python
# Sampling out 50 rows
sample_df = df.sample(n=50, random_state=123).reset_index(drop=True)
sample_df.head(50)
```

## 3.5 Generate output

```python
answers = []

for i, row in sample_df.iterrows():
    prompt = row["prompt"]
    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=[{"role": "user", "content": prompt}],
        max_completion_tokens=2000,
    )

    answer_text = response.choices[0].message.content.strip()
    answers.append(answer_text)
    print(f"Done: {i+1}/{len(sample_df)} | Preview: {answer_text[:30]}...")  # progress indicator & preview of answer
    
    # Safety break after 50 iterations (i = 0–49)
    if i == 49:
        break


# creates the final dataframe with a single column "answer"
answer_df = pd.DataFrame({"answer": answers})

answer_df.head(10)
```

### 3.5.1 Merge of input and output

```python
act_promt_answer_df = pd.concat([sample_df.reset_index(drop=True), answer_df], axis=1) # merging the two dfs
act_promt_answer_df.head(10)

act_promt_answer_df.to_csv("combined_input_output.csv", index=False)
```

## 4.1 Temperature
- The temperature parameter should controls how creative the model is in its responses. 
- Lower values (closer to 0) should result in more focused, deterministic outputs, while higher values (up to 2) produce more varied and creative results.

```python
from openai import OpenAI
client = OpenAI(api_key="INSERT-YOUR-API-KEY-HERE")

prompt = """
You're writing marketing copy for InertSport headphones. Your goal is to generate a persuasive product description including:
Strong inear hold
Active noise cancellation (ANC)
Foldable design
"""

# Create a request to the Chat Completions endpoint
response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[{"role": "user", "content": prompt}],
   
    max_completion_tokens=150,
    temperature=0
    
)

# Extract and print the response text
print(response.choices[0].message.content)
```

```python
prompt = """
You're writing marketing copy for InertSport headphones. Your goal is to generate a persuasive product description including:
Strong inear hold
Active noise cancellation (ANC)
Foldable design
"""

# Create a request to the Chat Completions endpoint
response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[{"role": "user", "content": prompt}],
   
    max_completion_tokens=150,
    temperature=2
    
)

# Extract and print the response text
print(response.choices[0].message.content)
```

#### Interpretation:
The results similar and even sharing simmilar structure and some identical beginnings of the sentence "Experience, Elevate".
This might be due to the quite detailed promt leaving only little room for variation for 150 tokens. 
To achieve more varied results, we can either raise "max_completion_tokens" or make the task description less specific.

#### Default value:
Temparuture defaults to 1 - https://platform.openai.com/docs/api-reference/chat/create

## 4.2 Top_p
- Is an alternative to temperature and works by considering only the tokens with top cumulative probabilities that add up to top_p
- A value of top_p=1 allows all possible tokens to be considered
- A smaller value limit the selection: top_p of 0.5 only considers the tokens, which probabilities cumulate up to 50% of

```python
prompt = """
Tell a bedtime story obout a pink flying unicorn
"""

# Create a request to the Chat Completions endpoint
response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[{"role": "user", "content": prompt}],
   
    max_completion_tokens=300,
    top_p= 0.95
    
)

# Extract and print the response text
print(response.choices[0].message.content)
```

Even sharing the name "Luna" and a similar plot of her flying and meeting someone.

## 4.3 Stop

Defines one or more stop sequences where the model stops the output.
The returned text will not contain the stop sequence.


```python
prompt = """
Create a financial plan for a small business that focuses on cost savings and long-term investments. Use the "Ω" symbol when you have explained the three most important points.
"""

# Create a request to the Chat Completions endpoint
response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[{"role": "user", "content": prompt}],
   
    max_completion_tokens=1000,
    stop=["Ω", "End"] # stops when "Ω" is encountered
    
)

# Extract and print the response text
print(response.choices[0].message.content)
```

#### Interpretation:
Allows users to control output length that model only generates to a certain mention of a sequence rather than a fixed token number.

#### Default:
Set to null

## 4.4 Presence_penalty

Can be set between -2.0 and 2.0. 
If positive, repeated mentions of tokens are penalized, if negative, they are rewarded.

```python
prompt = """
Invent a new everyday device that doesn’t exist yet. Explain what is is, who it is intended for, how it works in detail and why it makes life easier.
"""

# Create a request to the Chat Completions endpoint
response = client.chat.completions.create(
    model="gpt-4.1",
    messages=[{"role": "user", "content": prompt}],
   
    max_completion_tokens=250,
    presence_penalty = -2 # "negative penalty" rewarding frequent mentions 
    
)

# Extract and print the response text
print(response.choices[0].message.content)
```

```python
prompt = """
Invent a new everyday device that doesn’t exist yet. Explain what is is, who it is intended for, how it works in detail and why it makes life easier.
"""

# Create a request to the Chat Completions endpoint
response = client.chat.completions.create(
    model="gpt-4.1",
    messages=[{"role": "user", "content": prompt}],
   
    max_completion_tokens=250,
    presence_penalty = 2 # positive penalty, penalizing frequent mentions 
    
)

# Extract and print the response text
print(response.choices[0].message.content)
```

#### Interpretation:
When repeated mentions are rewarded, like in the first text (neg presence penalty), tokens like “dream” are more frequent frequently.
When repetitions are penalized.. like in the second text (pos presence penalty), the model avoids mentioning them again. It instead uses synonyms to describe the item.

#### Default:
Set to 0

## 4.5 Logit_bias



```python
prompt = """
Create a weekly meal plan for a student with a budget of €40. Include simple recipes, a grocery list and emergency to-go items.
"""

# Create a request to the Chat Completions endpoint
response = client.chat.completions.create(
    model="gpt-4o",
    messages=[{"role": "user", "content": prompt}],
   
    max_completion_tokens=500,
    logit_bias = {161877 : +13.5,
                 66 : + 16,
                 61381 : +6,
                 140188 : +14,
                 3232:+ 12 } 
    
)

# Extract and print the response text
print(response.choices[0].message.content)
```

```python
prompt = """
Create a weekly meal plan for a student with a budget of €40. Include simple recipes, a grocery list and emergency to-go items.
"""

# Create a request to the Chat Completions endpoint
response = client.chat.completions.create(
    model="gpt-4o",
    messages=[{"role": "user", "content": prompt}],
   
    max_completion_tokens=500,
    logit_bias = {161877 : +13.5,
                 66 : + 16,
                 61381 : +6,
                 140188 : +14,
                 3232:+ 12 },
    presence_penalty = 2 # added positive penalty, penalizing frequent mentions to combat loops
    
)

# Extract and print the response text
print(response.choices[0].message.content)
```

#### Interpretation:
When adjusting the settings so that tokens are way more likely than others, the model may start looping as seen in the first text.
In the second text the logit_bias is combined with a positive presence penalty which seemed to have avoided looping and certain tokes [140188, 3232] = [veget, ables] (that were favored) appear way more often in the second text.

#### Default:
Defaults to null
