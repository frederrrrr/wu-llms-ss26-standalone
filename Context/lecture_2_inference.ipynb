# Decoding

```python
prompt = 'students opened their'
```

```python
# 3) using automodel and generate
from transformers import AutoTokenizer, AutoModelForCausalLM

model = AutoModelForCausalLM.from_pretrained("gpt2-large")

tokenizer = AutoTokenizer.from_pretrained("gpt2-large")
model_inputs = tokenizer([prompt], return_tensors="pt")

generated_ids = model.generate(**model_inputs)
tokenizer.batch_decode(generated_ids)[0]
```

```python
generated_ids = model.generate(**model_inputs, max_new_tokens=2)
tokenizer.batch_decode(generated_ids)[0]
```

```python
generated_ids = model.generate(**model_inputs, num_beams=2, max_new_tokens=30)
tokenizer.batch_decode(generated_ids)[0]
```

```python
generated_ids = model.generate(**model_inputs, do_sample=True, top_k=4, max_new_tokens=2)
tokenizer.batch_decode(generated_ids)[0]
```

```python
generated_ids = model.generate(**model_inputs, do_sample=True, top_p=0.92, max_new_tokens=2)
tokenizer.batch_decode(generated_ids)[0]
```

```python
model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
model_name = "LiquidAI/LFM2-1.2B-Tool"

model = AutoModelForCausalLM.from_pretrained(model_name)

tokenizer = AutoTokenizer.from_pretrained(model_name)
model_inputs = tokenizer([prompt], return_tensors="pt")

generated_ids = model.generate(**model_inputs)
tokenizer.batch_decode(generated_ids)[0]
```
